```python
import pandas as pd
import numpy as np
df = pd.read_csv("files_for_lab/marketing_customer_analysis.csv")
df.head()
```

```python
#import necessary tools
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Reasponse rate by Sales Channel
sns.countplot(x = "Sales Channel", hue="Response", data=df)
plt.show()
```

```python
#Response rate by total claim amount
df_edit =df[df["Response"] == "Yes"]
sns.distplot(df_edit["Total Claim Amount"],  kde=False, label="Yes")
df_edit =df[df["Response"] == "No"]
sns.distplot(df_edit["Total Claim Amount"],  kde=False, label="No")

plt.legend()
plt.title("Response rate by total claim amount")
plt.xlabel("Total Claim Amount")
plt.ylabel("Density")

plt.show()
```

```python
#Response rate by Income
df_edit =df[df["Response"] == "Yes"]
sns.distplot(df_edit["Income"],  kde=False, label="Yes")
df_edit =df[df["Response"] == "No"]
sns.distplot(df_edit["Income"],  kde=False, label="No")

plt.legend()
plt.title("Response rate by Income")
plt.xlabel("Incom")
plt.ylabel("Density")

plt.show()
```
